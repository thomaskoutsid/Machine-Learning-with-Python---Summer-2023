# 12.

```python
import numpy as np

upper = correlations.where(np.triu(np.ones(correlations.shape), k = 1).astype(bool))

removed_corr = [column for column in upper.columns if any((upper[column] > 0.7) | (upper[column] < -0.7))]

correlations.drop(removed_corr, axis = 1, inplace = True)

# Resources used:
# https://stackoverflow.com/questions/29294983/how-to-calculate-correlation-between-all-columns-and-remove-highly-correlated-on
# https://www.andrews.edu/~calkins/math/edrm611/edrm05.htm
```

```python
removed_corr
```

The variables that had a magnitude of absolute(0.7) and over were removed. This was because +/- 0.7 and upward (to +/- 1.0) incidates a very high correlation. This means that the 7 columns listed above were dropped for this reason. If we look above at the correlations chart in 11a, we can see that these columns had strong correlations with others.
